```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from numpy import nan
```

**2. DATAnƒ± *Pandas* kitabxanasƒ± vasit…ôsi il…ô daxil edirik.**

```
DATA = pd.read_csv('/content/it_datacase.csv')
```

**3. *Numpy* and *Pandas* kitabxanalarƒ±ndan istifad…ô ed…ôr…ôk DATAmƒ±zda olan qadƒ±n, ki≈üi v…ô cinsi m…ôlum olmayan insanlarƒ±n sayƒ±nƒ± hesablayƒ±rƒ±q.** 

```
gender = (DATA.loc[:,'gender']).to_numpy()
female_c = np.count_nonzero(gender == 'Female')
male_c = np.count_nonzero(gender == 'Male')
unknown_c = np.count_nonzero(pd.isnull(gender))
known_c = len(gender) - unknown_c

def percentage(part, whole):
  return round(100 * float(part)/float(whole),1)

female_percentage = percentage(female_c, known_c)
male_percentage = percentage(male_c,known_c)


print(" Faylƒ±n ilkin variantƒ±nda qadƒ±nlarƒ±n sayƒ±:",female_c,"\n","Faylƒ±n ilkin variantƒ±nda ki≈üil…ôrin sayƒ±:",male_c,"\n","Cinsi m…ôlum olmayanlar:",unknown_c)

```

**4. Qadƒ±n v…ô ki≈üil…ôrin sayƒ±nƒ± m√ºqayis…ô etm…ôk √º√ß√ºn *pie chart*-dan istifad…ô edirik.** 

```
gender_numbers = np.array([female_c,male_c])  
mylabels = ["Qadƒ±nlar","Ki≈üil…ôr"]  
colors = ['lightcoral', 'lightskyblue']
explode = (0.05, 0.05)
total = female_c + male_c
def myfunction(x):
  return '{:.1f}%\n({:.0f})'.format(x,round(total*x/100))

plt.pie(gender_numbers, explode=explode, autopct=myfunction, colors=colors, shadow=True, startangle=90, wedgeprops = {'linewidth': 5}, rotatelabels=True) 
plt.legend(mylabels, loc="best")
plt.axis('equal')
plt.tight_layout()
plt.show()
```

**5. 51.3% qadƒ±nlar, 48.7% ki≈üil…ôr t…ô≈ükil etdiyi v…ô bu nisb…ôtin d…ôyi≈üm…ôm…ôsi √º√ß√ºn …ôlimizd…ô olan 55 cinsi m…ôlum olmayan insanƒ±n cinsl…ôrini eyni nisb…ôtd…ô b√∂l√ºr√ºk v…ô daha sonra yen…ô *pie chart* istifad…ô ed…ôr…ôk nisb…ôtin d…ôyi≈üm…ôdiyini g√∂r√ºr√ºk.** 

```
def percentage_number(number, perc):
  return round((number*perc)/100)

unknown_female = percentage_number(unknown_c,female_percentage)
unknown_male = percentage_number(unknown_c,male_percentage)

print(" T…ôxmin olunmu≈ü v…ô siyahƒ±ya …ôlav…ô olunmu≈ü qadƒ±nlarƒ±n sayƒ±:",unknown_female,"\n","T…ôxmin olunmu≈ü v…ô siyahƒ±ya …ôlav…ô olunmu≈ü  ki≈üil…ôrin sayƒ±:",unknown_male)
print("\n","\n")
gender = DATA['gender'].fillna('Female', limit = unknown_female)
gender = gender.fillna('Male').to_numpy()

female_c = np.count_nonzero(gender == 'Female')
male_c = np.count_nonzero(gender == 'Male')

total = female_c + male_c
plt.pie(gender_numbers, explode=explode, autopct=myfunction, colors=colors, shadow=True, startangle=90, wedgeprops = {'linewidth': 5}, rotatelabels=True) 
plt.legend(mylabels, loc="best")
plt.axis('equal')
plt.tight_layout()
print("\n","Cinsiyy…ôtl…ô baƒülƒ± olan s√ºtun emal olunduqdan sonraki v…ôziyy…ôtinin vizual nisb…ôti:","\n")
plt.show()
```
